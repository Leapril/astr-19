```python
#import matplotlib
import matplotlib.pyplot as plt
import numpy as np 
```

 Write cells that create an array x ranging from [0,1] in 100 steps and that defines a function that returns exp(x).

```python
#create an array x ranging from 0,1 in 100 steps 
x = np.linspace(0,1,100)

#defines a function that returns exp(x)
def exp(x):
    return np.exp(x)
    
```

In a new cell use the function to set y=exp(x), and then plot x vs. y. Label the x-axis as “Time [milliseconds]” and the y-axis as “Awesomeness”. Save the figure as a PDF.  

```python
#new cell use the function to set y = exp(x)
y = exp(x)

#plot the figure
#figsize() chnaged the size (width,height) of the graph figure
f = plt.figure(figsize=(5,5))
plt.plot(x,y)
#label x vs. y 
plt.xlabel('Time [millisexonds]')
plt.ylabel('Awesomeness')

#save the plot as a pdf
plt.savefig("session7-plot.pdf", bbox_inches="tight",facecolor="white")
```
