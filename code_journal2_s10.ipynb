On the numpy website, read about the random number distributions available for pulling random variates.  Select your favorite distribution (other than uniform) and pull 1000 random numbers from that distribution. Histogram the random numbers into 100 bins, and plot the histogram. Label your axes and save the figure as a PDF.


```python
#import inlines
import matplotlib.pyplot as plt
import numpy as np
```

```python
#pull 1000 random numbers between [0,1]
n = 1000 #number of data points to generate

#triangular distrubution
x = np.random.triangular(-3,0,8,n)

#create the histogram bin
width = 0.5 #width of the histogram

#plot the figure
plt.hist(x, bins=100, alpha=0.5, edgecolor="black")
plt.ylabel("N per bin")
plt.xlabel("x")
           
#save the figure as a pdf
plt.savefig("session9-plot.pdf", bbox_inches="tight",facecolor="white")
```
