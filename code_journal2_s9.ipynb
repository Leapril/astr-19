Use numpy to pull 1000 random numbers distributed uniformly between [0,1]. Histogram the random numbers into 100 bins, and plot the histogram. Label your axes and save the figure as a PDF.

```python
#import inlines
import matplotlib.pyplot as plt
import numpy as np
```

```python
#pull 1000 random numbers between [0,1]
n = 1000 #number of data points to generate
x = np.random.uniform(0,1n)#generate random data

#create the histogram bin
width = 0.5 #width of the histogram
histmin=np.floor(min(x))
histmax = np.ceil(max(x))+width

#bins = np.arange(histmin, histmax, width)
bins = 100

#plot the figure
plt.hist(x, bins=bins, alpha=0.5, edgecolor="black")
plt.ylabel("N per bin")
plt.xlabel("x")
           
#save the figure as a pdf
plt.savefig("session9-plot.pdf", bbox_inches="tight",facecolor="white")
```
